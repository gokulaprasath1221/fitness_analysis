```python
#group step_count 1-1000=1,1000-3000-2,3000-5000-3,above 5000 as 4
#df.loc[ df[“column_name”] == “some_value”, “column_name”] = “value”
#df.loc[df["gender"] == "male", "gender"] = 1
#df.loc[df['gender'] == 0, 'gender'] = gender_text[0]
#df.loc[df['gender'] == 1, 'gender'] = gender_text[1]
#df.loc[df['gender'] == 2, 'gender'] = gender_text[2]
#erroe note bracket should be behind data frame
df4.loc[df4["step_count"] <= 1000, "step_count"] = 1
df4.loc[(df4["step_count"] > 1000) & (df4["step_count"] <= 3000), "step_count"] = 2
df4.loc[(df4["step_count"] > 3000) & (df4["step_count"] <= 5000), "step_count"] = 3
df4.loc[df4["step_count"] > 5000, "step_count"] = 4
```

```python
df4
```

```python

df4.isnull().sum()
```

```python
df4["step_count"].value_counts()
```

```python
#df.groupby('StationID').mean()
# df.groupby('StationID')['BiasTemp'].mean()
df5=df4.groupby("step_count")["calories_burned"].mean()
```

```python
df5
```

```python
#df5 = df5.to_frame().reset_index(drop=True)
df5 = df5.to_frame()
```

```python

plt.figure(figsize=(9,6))
label=("calories_burned",)

plt.bar(x=df5['step_count'],height=df5['calories_burned'],width=0.5,color="orange")
plt.xlabel("step_count",fontsize=12,weight="bold")
plt.ylabel("calories burned",fontsize=12,weight="bold")
plt.title("Step_Count VS Calories_Burned",fontsize=16,weight="bold")
plt.legend(labels=label,loc=2)
plt.xticks([0,1,2,3,4])
plt.show()
```
